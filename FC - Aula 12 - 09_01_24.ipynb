````python
import math

def primes(n):

    #cria uma lista n valores True
    isprime = [True]*(n+1)

    #lista para armazenar os primos
    lprimes = []

    # marca 0 e 1 como não primos
    isprime[0] = isprime[1] = False

    # começa com 2, o primeiro primo
    isprime[2] = True

    for i, prime in enumerate(isprime):
        if prime:
            #salvamos o primo e marcamos os
            #múltiplos dele como não primos
            lprimes.append(i)
            #varre os múltiplos de i
            for m in range(i, (n+1), i):
                isprime[m] = False
    return lprimes
````

````python
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
````

````python
def pacote_gaussiano(omega=1, sigma=1, mu=0, cor='red'):
    plt.figure(1)
    x = np.linspace(-10, 10, num=1000)
    y = np.sin(omega*x)*np.exp(-(x-mu)**2/(2*sigma**2))
    plt.plot(x, y,color=cor)
    plt.ylim(-1, 1)
    plt.show()
````

````python
pacote_gaussiano(omega=10,mu=-7.5)
````

````python
interactive_plot = interactive(pacote_gaussiano,
                               omega=(1, 100,1),
                               mu=(0, 10, 1),
                               sigma=(0.1,2,0.1),
                               cor=['red','blue','gray'])
interactive_plot
````

````python
range_problemas_caps4 = [
    ["P4.1.1", "P4.1.3"],
    ["P4.2.1", "P4.2.5"],
    ["P4.3.1", "P4.3.6"],
    ["P4.4.1", "P4.4.4"],
    ["P4.6.1", "P4.6.5"],
]
````

````python
lista_problemas = []
for range_p in range_problemas_caps4:
    i_min=int(range_p[0].split('.')[-1])
    i_max=int(range_p[1].split('.')[-1])

    prefix = '.'.join(range_p[0].split('.')[:2])

    for i in range(i_min,i_max+1):
        lista_problemas.append(f"{prefix}.{i}")
````

````python
print(f"Total de problemas nos capítulos 4: {len(lista_problemas)}")
````

````python
import numpy as np

n_problemas = 5

problemas_estudantes = {}

for estudante in estutandes_pos+estudantes_grad:
    problemas_sel = np.random.choice(lista_problemas,size=n_problemas,replace=False)
    problemas_estudantes[estudante.title()] = sorted(list(problemas_sel))
````
