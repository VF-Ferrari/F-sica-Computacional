````python
# Initial value of y at t=0, lifetime in s
N, tau = 10000, 28
# Maximum time to consider (s)
tmax = 100
# A suitable grid of time points, and the exponential decay itself
t = np.linspace(0, tmax, 1000)
y = N * np.exp(-t/tau)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(t, y)

# The number of lifetimes that fall within the plotted time interval
ntau = tmax // tau + 1
# xticks at 0, tau, 2*tau, ..., ntau*tau;
#yticks at the corresponding y-values
xticks = [i*tau for i in range(ntau)]
yticks = [N * np.exp(-i) for i in range(ntau)]

print(xticks, yticks)

ax.set_xticks(xticks)
ax.set_yticks(yticks)

# xtick labels: 0, tau, 2tau, ...
xtick_labels = [r'$0$', r'$\tau$'] + [r'${}\tau$'.format(k) for k in range(2,ntau)]
ax.set_xticklabels(xtick_labels)
# corresponding ytick labels: N, N/e, N/2e, ...
ytick_labels = [r'$N$',r'$N/e$'] + [r'$N/{}e$'.format(k) for k in range(2,ntau)]
ax.set_yticklabels(ytick_labels)

ax.set_xlabel(r'$t\;/\mathrm{s}$')
ax.set_ylabel(r'$y$')
ax.grid()
````

````python
# A selection of functions on rn abcissa points for 0 <= x < 1
rn = 100
rx = np.linspace(0, 1, rn, endpoint=False)

def tophat(rx):
    """ Top hat function: y = 1 for x < 0.5, y=0 for x >= 0.5 """
    ry = np.ones(rn)
    ry[rx>=0.5]=0
    return ry

# A dictionary of functions to choose from
ry = {'half-sawtooth': lambda rx: rx.copy(),
      'top-hat': tophat,
      'sawtooth': lambda rx: 2 * np.abs(rx-0.5)}

# Repeat the chosen function nrep times
nrep = 4
x = np.linspace(0, nrep, nrep*rn, endpoint=False)
y = np.tile(ry['top-hat'](rx), nrep)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y, 'k', lw=2)

# Add a bit of padding around the plotted line to aid visualization
ax.set_ylim(-0.1,1.1)
ax.set_xlim(x[0]-0.5, x[-1]+0.5)

# Customize the tick marks and turn the grid on
ax.minorticks_on()
ax.tick_params(which='major', length=10, width=2, direction='inout')
ax.tick_params(which='minor', length=5, width=2, direction='in')
ax.grid(which='both')
````

### Barras de erro

Para fazer gráficos com barras de erro usamos a função `ax.errorbar`. Essa função permite especificar os erros tanto na coordenada $x$ (`xerr`) quanto na coordenada $y$ (`yerr`):

- `None`: sem barras de erro;
- `yerr=escalar`: todos os pontos terão o mesmo erro ($\pm$ escalar);
- `yerr=array`: um array ou lista com shape (número de pontos, 1). Nesse caso, cada elemento se refere ao erro do $i$-ésimo ponto;
- `yerr=array`: um array ou lista com shape (número de pontos, 2), ou seja, duas colunas. Nesse caso, as colunas contêm os erros para cima e para baixo referente a cada ponto.


A tabela abaixo mostra os outros argumentos da função:

image.png

````python
f, ax = plt.subplots(figsize=(8,5))

x = np.array([0.3, 0.5, 0.7, 0.9])
y = np.array([ 1. , 2. , 2.5, 3.9])

yerr = np.array([0.1, 0.15, 0.2, 0.0]) # erro simétrico
xerr = 0.05 # erro constante

ax.errorbar(x, y, yerr, xerr, fmt='o', ls='', capsize=5)
````

Vamos supor um modelo em que:

$$
\text{carga} = A \exp(-B t)
$$

Linearizando, temos:

$$
\log(\text{carga}) = \log(A) -B\, t
$$

*Exemplo E7.7:* Difusão de calor em uma barra de metal muito comprida. Resolvendo a equação de difusão:

$$
\frac{\partial\theta}{\partial t} = D \frac{\partial^2\theta}{\partial x^2}
$$

com $D$ sendo a difusividade térmica, temos que a temperatura $\theta(x,t)$ é dada por:

$$
\theta(x,t) = \frac{H}{c_p A} \frac{1}{\sqrt{4\pi D t}}\exp \left(-\frac{x^2}{4 D t}\right)
$$

sendo que $c_p$ é o calor específico, $A$ é área da seção transversal da barra e $H$ e taxa de calor transferido para o centro da barra.

O código abaixo faz o gráfico dessa solução para 3 tempos diferentes considerando dois metais com diferentes difusividades.

````python
# Parâmetros
A, H = 1.e-4, 1.e3

#temperatura inicial
theta0 = 300

Cu_D,Cu_cp = 1.11e-4, 3.45e7
Fe_D,Fe_cp = 2.30e-5, 3.50e7
````

````python
def solucao(t, cp, D, A=A, H=H):
    xlim, nx = 0.05, 1000
    x = np.linspace(-xlim, xlim, nx)

    theta = theta0 + (H/(cp*A))*(1/np.sqrt(D*t * 4*np.pi))* np.exp(-x**2/(4*D*t))

    return x*100,theta
````

````python
f, ax = plt.subplots(nrows=3,ncols=2,figsize=(6,8),sharey=True,sharex=True)

tempos = (1e-2, 0.1, 1)

for i,ax_ in enumerate(ax[:,0]):
    ax_.plot(*solucao(tempos[i],Cu_cp,Cu_D))
    ax_.set_title(fr'Cu, $t={tempos[i]:.2f}$ s')
    ax_.set_xlabel('Posição $x$ (cm)')
    ax_.set_ylabel(r'Temperatura $\theta$ (K)')

for i,ax_ in enumerate(ax[:,1]):
    ax_.plot(*solucao(tempos[i],Fe_cp,Fe_D))
    ax_.set_title(fr'Fe, $t={tempos[i]:.2f}$ s')
    ax_.set_xlabel('Posição $x$ (cm)')
    ax_.set_ylabel(r'Temperatura $\theta$ (K)')

f.tight_layout()
````

*Exemplo E7.16*: Gráfico representando o espectro eletromagnético.

````python
# range de comprimentos de onda (nm)
lmin, lmax = 250, 1000
x = np.linspace(lmin, lmax, 1000)

# função seno com comprimento de onda variando
# o -1 é para ficar em ordem crescente (lmax+lmin-x)
wv = (np.sin(10 * np.pi * x / (lmax+lmin-x)))[::-1]

f, ax = plt.subplots(figsize=(8,5),
                     subplot_kw=dict(facecolor="#505050"))

ax.plot(x, wv, c='w', lw=2)
ax.set_xlim(250,1000)
ax.set_ylim(-2,2)

#texto UV
ax.text(310, 1.5, 'UV', color='w', fontdict={'fontsize': 20})

#texto e seta visível
ax.text(530, 1.5, 'Visível', color='k', fontdict={'fontsize': 20})
ax.annotate('', (400, 1.3), (750, 1.3), arrowprops={'arrowstyle': '<|-|>',
                                                    'color': 'w', 'lw': 2})
#texto do IR
ax.text(860, 1.5, 'IR', color='w', fontdict={'fontsize': 20})

#linhas verticais delimitando o visível
ax.axvline(400, -2, 2, c='w', ls='--')
ax.axvline(750, -2, 2, c='w', ls='--')

#desliga os ticks ao longo do eixo y
ax.yaxis.set_visible(False)

ax.set_xlabel(r'$\lambda$ (nm)')

#define as cores para o visível
rainbow_rgb = {(400, 440): '#8b00ff',
               (440, 460): '#4b0082',
               (460, 500): '#0000ff',
               (500, 570): '#00ff00',
               (570, 590): '#ffff00',
               (590, 620): '#ff7f00',
               (620, 750): '#ff0000'}

for wv_range, rgb in rainbow_rgb.items():
    #desenha os retângulos coloridos
    ax.axvspan(*wv_range, color=rgb, ec='none', alpha=1)
````

````python
FEMALE, MALE = 0, 1
dt = np.dtype([('mass', 'f8'), ('height', 'f8'), ('gender', 'i2')])
data = np.loadtxt('body.dat.txt', usecols=(22,23,24), dtype=dt)
````

````python
def get_cov_ellipse(cov, centre, nstd, **kwargs):
    """
    Return a matplotlib Ellipse patch representing the covariance matrix
    cov centred at centre and scaled by the factor nstd.

    """

    # Find and sort eigenvalues and eigenvectors into descending order
    eigvals, eigvecs = np.linalg.eigh(cov)
    order = eigvals.argsort()[::-1]
    eigvals, eigvecs = eigvals[order], eigvecs[:, order]

    # The anti-clockwise angle to rotate our ellipse by
    vx, vy = eigvecs[:,0][0], eigvecs[:,0][1]
    theta = np.arctan2(vy, vx)

    # Width and height of ellipse to draw
    width, height = 2 * nstd * np.sqrt(eigvals)
    return Ellipse(xy=centre, width=width, height=height,
                   angle=np.degrees(theta), **kwargs)
````

````python
fig, ax = plt.subplots(figsize=(8,5))

labels, colours = ['Female', 'Male'], ['#762a83', '#1b7837']

for gender in (FEMALE, MALE):
    sdata = data[data['gender']==gender]
    height_mean = np.mean(sdata['height'])
    mass_mean = np.mean(sdata['mass'])
    cov = np.cov(sdata['mass'], sdata['height'])
    ax.scatter(sdata['height'], sdata['mass'], color=colours[gender],
               label=labels[gender], s=3)
    e = get_cov_ellipse(cov, (height_mean, mass_mean), 3,
                        fc=colours[gender], alpha=0.4)
    ax.add_artist(e)

ax.set_xlim(140, 210)
ax.set_ylim(30, 120)
ax.set_xlabel('Altura (cm)')
ax.set_ylabel('Massa (kg)')
ax.legend(loc='upper left', scatterpoints=1)
````
