*Exemplo E2.3* [Fórmula de Heron](https://en.wikipedia.org/wiki/Heron%27s_formula#Proofs) para área de um triângulo com lados $a, b, c$:

$$A = \sqrt{s(s-a)(s-b)(s-c)}$$

com $s=(a+b+c)/2$.


````python
import math

a = 4.503
b = 2.377
c = 3.902

s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))

area
````

````python
#Vamos ver funções mais adiante, mas podemos fazer aqui também.
def calc_area(a, b, c):
    """
    Essa função aplica a fórmula de Heron.

    a: Real
    b: Real
    c: Real
    ----
    retorna a área de um triângulo de lados a,b,c.

    """
    s = (a + b + c) / 2
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    return(area)
````

````python
calc_area(4.503,2.377,3.902)
````

````python
calc_area() #mostrar a doc string.
````

## **Imutabilidade e identidade**

Os objetos que vimos até agora são imutáveis. Isso significa que, uma vez criados, eles não podem ser modificados.

Por exemplo:

````python
a = 8
b = a
````

A primeira linha cria o objeto com valor 8 na memória e atribui a ele o nome de *a*. A segunda linha atribui o nome *b* **para o mesmo objeto**.

Veja que as ids deles são iguais:

````python
id(a), id(b)
````

Ao atribuir um novo valor para *a*, na verdade, estamos criando um novo objeto.

Veja:

````python
a = 3.14
a
````

````python
id(a), id(b)
````

Note que a id de *a* mudou enquanto a de $b$ permaneceu igual. A Figura ilustra isso melhor:

image.png

Podemos usar operador 'is' para verificar se duas variáveis apontam para um mesmo objeto:

````python
a = 2432
b = a
a is b
````

Note que o 'is' não se refere ao valor da variável, mas sim a 'identidade' do objeto que elas representam.

Veja abaixo:

````python
c = 2432
c is a
````

````python
c == a
````

Assim, por exemplo, em:

````python
a = 800
print(a,id(a))

a = a + 1
print(a,id(a))
````

A primeira linha cria um objeto inteiro 800 e atribui a ele o nome de *a*. A linha 4, cria um novo objeto com valor 801 (a soma é calculada primeiro) e atribui a ele o mesmo nome *a*. Nessa segunda atribuição, o valor 800 será esquecido, a menos que exista algum outro objeto que se refira a ele. Esse processo é conhecido como [**garbage collection**](https://en.wikipedia.org/wiki/Garbage_collection_(computer_science)).


O processo de retribuir valores a variáveis é muito comum e por isso existem algumas maneiras mais compactas para isso. Por exemplo,

````python
a = 1
a = a + 5
print(a)
````

````python
a = 1
a += 5
print(a)
````

````python
a = 2
a *= 3
print(a)
````

````python
a = 2
a /= 3
print(a)
````

**Formatando Strings**

Usando o método '.format':



````python
'{} plus {} equals {}'.format(2, 3, 'five ')
````

No código acima, os "{}" indicam posições que serão substituídas pelo método *format* em ordem sequencial.

Também é possível usar nomes:

````python
'{num1} plus {num2} equals {answer}'.format(num1=2, num2=3, answer ='five')
````

É possível especificar um tamanho mínimo:

````python
'=== {} === '.format ('Python')
````

Por padrão, o alinhamento é à esquerda. Mas isso pode ser modificado com:
* '<': alinhar à esquerda (default);
* '>': alinhar à direita (default);
* 'ˆ': centralizar.

````python
'=== {0:<12} === '.format('Python')
````

````python
'=== {0:>12} === '.format('Python')
````

````python
'=== {0:^12} === '.format('Python')
````

Para incluir o char "{" ou "}", use "{{" ou "}}".

````python
'==={{{0:^12}}}=== '.format('Python')
````
