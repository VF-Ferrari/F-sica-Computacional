## **Tipos numéricos**

O Python tem três tipos numéricos usados para representar números:



*   Inteiro: *int*
*   Ponto flutuante: *float*
*   Complexo: *complex*

**Inteiros**

Exemplos de *int* são 1, 8, -72 e 3847298893721407. Aritmética com inteiros é exata em Python e não existe limite para representar inteiros (apenas o limite de memória do PC).

Você pode usar o caractere "_" como separador de milhar:



````python
a = 299792458
b = 299_792_458

print(a,b)
````

**Ponto flutuante** (*float*)

Exemplos de *float* são 1.2, -0.36 e $1.67263 \times 10^{−7}$. *Floats* são representados em forma binária de maneira *aproximada*. A precisão usual é de 15-16 casas decimais. **Aritmética com *floats* é sempre aproximada**, mas é "boa o suficiente" para a maioria das aplicações científicas.


````python
a = 4/3 #1.333333...
f"{a:.20f}"
````

````python
a = 0.01
b = 0.1**2

a==b
````

````python
f"{a:.20f}"
````

````python
f"{b:.20f}"
````

Qualquer número contendo um "." é interpretado como *float*:

````python
a = 1.
type(a) #type é uma função que retorna o tipo da variável
````

Podemos usar notação científica com o caractere "e" ou "E":

````python
carga_eletron = -1.6e-19
print(carga_eletron)
````

````python
carga_eletron = -1.602_176_634E-19 #também podemos usar "_"
print(carga_eletron)
````

**Números complexos**

Em Python, $j$ é unidade imaginária e número complexos são representados usando dois *floats*:

$z = a+b\,j$

com $a$ e $b$ sendo *floats*. Portanto, aritmética com *complex* estão sujeitas às mesmas limitações dos *floats*.

Podemos definir números complexos de duas maneiras:



````python
z = 5+3j
print(z)
````

````python
z = complex(5,3) #complex("parte real", "parte imaginária")
print(z)
````

## **Funções matemáticas**

'Nativamente' o Python oferece apenas duas funções matemáticas:

*   *abs*: retorna o valor absoluto de um número;
*   *round*: arredonda um número.

````python
abs(-5.2)
````

````python
abs(-2)
````

````python
abs(3+4j) #raiz quadrada de 9+16=25
````

O exemplo anterior é o que chamamos de polimorfismo, no qual uma mesma função pode retornar o valor absoluto ($|x|$) ou módulo ($x=\sqrt{\text{PARTE_RE}^2+\text{PARTE_IM}^2}$) de um número complexo.

````python
round(-9.62)
````

````python
round(7.5)
````

````python
round(4.5)
````

Por padrão, a função *round* arredonda para o inteiro par mais próximo. O chamado "banker's rounding" ou arredondamento de banqueiro.

Podemos especificar o número casas decimais passando um *parâmetro* para a função *round*:

````python
round(3.141592653589793, 3)
````

````python
round(96485.33289, 0)
````

````python
round(96485.33289, -1)
````
